<h1>Задание 1</h1>

На маркетплейсе есть огромный объем данных содержащий записи о продажах различных продуктов. Наша цель — рассчитать общий доход, полученный от каждого продукта за определённый период. 

Необходимо решить задачу через классический python, numpy и numpy с использованием векторизации


Данные: 
products = ['автомобиль HAVAL', 'обувь', 'одежда', 'телефоны'] (при решении через numpy необходимо представить это в виде np.array() )
prices = [20000, 3000, 45600, 5000] (при решении через numpy необходимо представить это в виде np.array() ) 

Данные о продажах необходимо сгенерировать самостоятельно и вывести в одном параграфов в ноутбуке, ниже приведен пример генерации случайных значений 

```python
import random
import time
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = random.Random(0)
variant = 15 # подставьте номер своего  варианта
products = ['автомобиль HAVAL', 'обувь', 'одежда', 'телефоны']
prices = [20000, 3000, 45600, 5000]
sales = [[rng.randint(0, variant+100) for i in range(100000)] 
         for i in range(len(products))]
print(sales[0])
```

**классический python**

```python
# пример генерации на python
start_time_python = time.time()

product_income_python = [sum(sales[i])*prices[i] 
                  for i in range(len(prices))]

end_time_python = time.time() - start_time_python

print('Сумма продаж за период\n')
for i in range(len(products)):
    print(products[i], ':', product_income_python[i], sep=' ')
print('\nВремя выполнения %s секунд\n' % end_time_python)
```

**numpy**

```python
products = np.array(products)
sales = np.array(sales, dtype=np.int64)
prices = np.array(prices, dtype=np.int64)

start_time_numpy = time.time()

product_income_numpy = np.array([np.sum(sales[i])*prices[i] 
                  for i in range(len(prices))])

end_time_numpy = time.time() - start_time_numpy

print('Сумма продаж за период\n')
for i in range(len(products)):
    print(products[i], ':', product_income_numpy[i], sep=' ')
print('\nВремя выполнения %s секунд\n' % end_time_numpy)
```

**numpy multiply**

```python
start_time_numpy_v = time.time()

product_income_numpy_v = np.multiply(np.sum(sales, axis=1), prices)

end_time_numpy_v = time.time() - start_time_numpy_v

print('Сумма продаж за период\n')
for i in range(len(products)):
    print(products[i], ':', product_income_numpy_v[i], sep=' ')
print('\nВремя выполнения %s секунд\n' % end_time_numpy_v)
```

```python
for i in range(len(products)):
    if (product_income_python[i] == product_income_numpy[i]) and \
        (product_income_numpy[i] == product_income_numpy_v[i]):
        print('для продукта', products[i], 'суммы совпадают')
    else:
        print('!! для продукта', products[i], 'суммы не совпадают')

print('\nВремя выполнения\n')
print('python:\t', end_time_python)
print('numpy:\t', end_time_numpy)
print('numpy_v:', end_time_numpy_v)
```

Данные в различных реализациях должны совпадать.

После генерации необходимо сравнить доход от каждого продукта и время на выполнение.

Необходимо написать с использованием numpy реализацию одного из следующих алгоритмов: 

- heapsort (вариант: 1,5,9,13,17,21)

- mergesort (вариант: 2,6,10,14,18,22)

- stable (вариант: 3,7,11,15,19)

- quicksort (вариант: 4,8,12,16,20) 

Сравнить скорость выполнения этих алгоритмов на чистом python, вашей реализации и numpy.sort() 

У меня вариант 15, stable sort. Я буду использовать сортировку вставками

```python
def insertion_sort(array):
    n = len(array)
    for i in range(1, n):
        j = i
        tmp = array[i]
        while j > 0 and tmp < array[j - 1]:
            array[j] = array[j - 1]
            j -= 1
        array[j] = tmp
```

```python
np.random.seed(1)
array = np.random.randint(low=0, high=1000, size=20000)
array_list = array.tolist()
print(array)
```

У нас есть Полный каталог всех случаев, когда кто-то проклинал или истекал кровью в фильме Квентина Тарантино. https://github.com/fivethirtyeight/data/tree/master/tarantino

Давайте представим эти данные в виде структурированных массивов и определим:

- число проклятий, летальных исходов и их отношение по фильмам 

- частота употребления конкретных прокрятий 

- распределение времени между проклятиями 

```python
data = np.genfromtxt('tarantino.csv', dtype = 'U20,'*3 + 'float', delimiter = ",", names=True)
```

```python
data
```

```python
data_grouped = np.split(data, np.sort(np.unique(data['movie'], return_index=True)[1])[1:])
```

```python
for arr in data_grouped:
    num_of_words = arr[arr['type'] == 'word'].shape[0]
    num_of_deaths = arr[arr['type'] == 'death'].shape[0]
    print('Фильм:', arr['movie'][0])
    print('Число проклятий:', num_of_words)
    print('Число смертей:', num_of_deaths)
    print('Отношение числа проклятий к числу смертей:', np.round(num_of_words/num_of_deaths, 2))
    print('\n')
```

Теперь найдем частоту употребления конкретных проклятий

```python
words, counts = np.unique(data['word'], return_counts=True)
```

```python
word_count = dict(zip(words, counts))
del word_count['']
```

```python
word_count
```

```python
plt.figure(figsize=(20,20))
plt.xticks(rotation=90)
plt.bar(word_count.keys(), height=word_count.values())
```

А теперь изучим распределение времени между проклятиями

```python
data_grouped = data[data['type'] == 'word']
data_grouped = np.split(data_grouped, np.sort(np.unique(data_grouped['movie'], return_index=True)[1])[1:])
```

```python
time_diff = np.array([])
for arr in data_grouped:
    time_diff = np.append(time_diff, np.diff(np.sort(arr['minutes_in'])))
```

```python
fig, axs = plt.subplots(2, figsize=(15,15))
axs[0].hist(time_diff, bins=50)
axs[0].title.set_text('Distribution')
axs[1].hist(time_diff, bins=50, log=True)
axs[1].title.set_text('Distribution (log scale)')
```
